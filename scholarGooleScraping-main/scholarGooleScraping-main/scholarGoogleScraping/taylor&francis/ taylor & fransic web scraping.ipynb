```python
from bs4 import BeautifulSoup
import requests
import pandas
titlesArray=[]
keywordsArray=[]
affilisArray=[]
abstractsArray=[]
publicationsArray=[]
linksArray=[]
publisherArray=[]

tags = requests.get('https://scholar.google.com/scholar?start=40&q=blockchain+iran+Taylor+%26+Francis+-BOOK&hl=en&as_sdt=0,5&as_ylo=2018').text
soup = BeautifulSoup(tags, 'html') 


                            
linksOfTitles=soup.select('#gs_res_ccl_mid > div > div.gs_ri > h3>a')
for link in linksOfTitles:
  linksArray.append(link['href'])
        
data={ 'links':linksArray}
d= pandas.DataFrame(data)
d[d.links.str.contains('link.springer.com')]
# print(d[d.links.str.contains('link.springer.com')])

for ind in d[d.links.str.contains('tandfonline.com')].index:
    link2=d[d.links.str.contains('tandfonline.com')]['links'][ind]
    print(link2)
    tags2 = requests.get(link2)
    soup2 = BeautifulSoup(tags2.text, 'html')
    abstracts=soup2.select(' div.hlFld-Abstract> div> p') 
       
      
    publisherArray.append('taylor&fransic')

    
    for abstract in abstracts:
     if abstract.text!='ABSTRACT'and abstract.text!='Abstract':
      abstractsArray.append(abstract.text)

    titles=soup2.select('#fa57727f-b942-4eb8-9ed2-ecfe11ac03f5 > div > div > h1 > span')
    for title in titles:
     titlesArray.append(title.text)

    affiliations=soup2.select('#fa57727f-b942-4eb8-9ed2-ecfe11ac03f5 > div > div > div.literatumAuthors > div > div > span')
    for affili in affiliations:
     affilisArray.append(affili)
        
    
data={ 'publisher':publisherArray,'abstracts':abstractsArray, 'titles':titlesArray,'affiliations':affilisArray, }
df= pandas.DataFrame(data)
print(df)
df.to_excel("C:/Users/musta/Desktop/scholar_scrape/excel/taylor4.xlsx")
```
