```python
from bs4 import BeautifulSoup
import requests
import pandas
titlesArray=[]
keywordsArray=[]
affilisArray=[]
abstractsArray=[]
publicationsArray=[]
linksArray=[]
publisherArray=[]

tags = requests.get('https://scholar.google.com/scholar?start=0&q=+blockchain+iran+elsevier+-BOOK&hl=en&as_sdt=0,5&as_ylo=2018').text
soup = BeautifulSoup(tags, 'html') 


                            
linksOfTitles=soup.select('#gs_res_ccl_mid > div > div.gs_ri > h3>a')
for link in linksOfTitles:
  linksArray.append(link['href'])
        
data={ 'links':linksArray}
d= pandas.DataFrame(data)
d[d.links.str.contains('sciencedirect.com')]
# print(d[d.links.str.contains('link.springer.com')])

for ind in d[d.links.str.contains('sciencedirect.com/')].index:
    link2=d[d.links.str.contains('sciencedirect.com')]['links'][ind]
#     print(link2)
    pii=link2.split('/')[-1]
    link22='https://api.elsevier.com/content/article/pii/'+str(pii)+'?APIKey=776ca8dde2af1ea9947c24b6500168de'
    
    
    
    
    
    
    
    tags2 = requests.get(link22)
    soup2 = BeautifulSoup(tags2.text, 'html')
    
    
    openaccess=soup2.select_one('coredata >openaccess')
    abstracts=openaccess.find_previous()
    abstractsArray.append(abstracts.text)


               

    coredata=soup2.select_one('coredata')
    titles=coredata.contents[5]
    titlesArray.append(titles.text)
    
    publisherArray.append('elsevier')

#     affiliations=soup2.select('#author-information-content > ol')
#     for affili in affiliations:
#      affilisArray.append(affili)
        
          
data={'publisher':publisherArray, 'abstracts':abstractsArray, 'titles':titlesArray}
df= pandas.DataFrame(data)
df = df.replace({"^\s*|\s*$": ""}, regex=True)
print(df)
df.to_excel("C:/Users/musta/Desktop/scholar_scrape/excel/elsevier0.xlsx")
```
